```python
import pandas as pd
from serpapi import search
from time import sleep

# Read Excel file and generate search queries
df = pd.read_excel(r'Websites_for_businesses.xlsx')

# Define your SerpApi API key
api_key = "42c0348e91eae86292a118aa31aa1a15f0b7f8090e44d8d853c97206f88c45a0"

# List to store the results
results_list = []

# Process the first 20 rows
for _, row in df.head(20).iterrows():
    query = f"{row['Provider name']} {row['Address 1']} nursing home site:co.uk"

    # Make the request to SerpApi
    params = {"engine": "google", "q": query, "api_key": api_key}
    results = search(params)

    # Sleep to avoid hitting rate limits
    sleep(2)

    # Extract relevant information from search results
    search_results = results.get('organic_results', [])
    filtered_links = [result['link'] for result in search_results if any(keyword in result.get('link', '').lower() for keyword in ['nursing', 'care', 'home'])]

    # Create a dictionary with the relevant information
    result_dict = {
        'Google_search_query': query,
        'Official_website': filtered_links[0] if filtered_links else 'Not found'
    }

    # Append the dictionary to the results list
    results_list.append(result_dict)

# Create a DataFrame from the results list
results_df = pd.DataFrame(results_list)

# Merge the results DataFrame with the original DataFrame on index
merged_df = pd.concat([df.head(20), results_df], axis=1)

# Save the merged DataFrame back to the Excel file
merged_df.to_excel(r'Websites_for_businesses_updated.xlsx', index=False)

# Display the final merged DataFrame
print(merged_df)
```
